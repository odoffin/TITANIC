## Preparing Data for Training

```python
titanic
```

```python
titanic.shape
```

```python
sns.heatmap(titanic.isnull(), yticklabels = False, cmap = 'Blues')
```

```python
titanic.drop(['Cabin', 'Name', 'Ticket', 'Embarked'], axis = 1, inplace = True)
```

```python
titanic
```

```python
sns.heatmap(titanic.isnull(), yticklabels = False, cmap = 'Blues')
```

```python
plt.figure(figsize = (15,10))
sns.boxplot(x = 'Sex', y  = 'Age', data = titanic)
```

```python
#Future engineering
def input_age(cols):
    Age = cols[0]
    Sex = cols[1]
    
    if pd.isnull(Age):
        if Sex == 'male':
            return 29
        else:
            return 25
    else:
        return Age 
        
```

```python
titanic['Age'] = titanic [['Age', 'Sex']].apply(input_age, axis = 1)
```

```python
sns.heatmap(titanic.isnull(), yticklabels = False, cmap = 'Blues')
```

```python
titanic['Age'].hist(bins = 20)
```

```python
titanic.drop(['PassengerId'], axis = 1, inplace = True)
```

```python
titanic
```

```python
male = pd.get_dummies(titanic['Sex'], drop_first = True)
male
```

```python
titanic.drop(['Sex'], axis = 1, inplace = True)
```

```python
titanic
```

```python
titanic = pd.concat([titanic, male], axis = 1)
```

```python
titanic
```

```python
X = titanic.drop('Survived', axis = 1).values
```

```python
X
```

```python
y = titanic['Survived'].values
```

```python
y
```

## Model Training

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 10)
```

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(random_state = 0)
classifier.fit(X_train, y_train)
```

## Model Testing

```python
y_predict = classifier.predict(X_test)
y_predict
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_predict)
sns.heatmap(cm, annot = True, fmt='d')
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test, y_predict))
```

```python
classifier.score(X_test, y_test)
```
